```python
cursor = conn.cursor(buffered=True)

query9 = """SELECT  
    corsi.id_corsi, 
    corsi.nome_corso, 
    corsi.numero_max, 
    COUNT(studenti_corsi.id) AS totale_studenti
FROM studenti_corsi
INNER JOIN corsi 
    ON corsi.id_corsi = studenti_corsi.id_corso
GROUP BY 
    corsi.id_corsi, 
    corsi.nome_corso, 
    corsi.numero_max
    having totale_studenti >= corsi.numero_max;"""

cursor.execute(query9)
group = cursor.fetchall()

for id_corsi, nome_corso, numero_max, totale_studenti in group:
    print(id_corsi, nome_corso, f"limite {numero_max}", f" totale stud {totale_studenti}")
```

```python
query10= """SELECT 
    studenti.nome, 
    studenti.cognome,
    COUNT(studenti_corsi.id) AS corsi
FROM studenti
INNER JOIN studenti_corsi 
    ON studenti.id_studenti = studenti_corsi.id_studenti
GROUP BY 
    studenti.id_studenti, 
    studenti.nome, 
    studenti.cognome
"""
    
cursor.execute(query10)
part = cursor.fetchall()

for nome, cognome, corsi in part:
    print(nome, cognome, corsi)

```

```python
query13 = """ 
SELECT 
    docenti.id_insegnante, 
    docenti.nome, 
    docenti.cognome,
    COUNT(corsi.id_corsi) AS numero_corsi_prof
FROM docenti
LEFT JOIN corsi 
    ON docenti.id_insegnante = corsi.id_insegnate
GROUP BY 
    docenti.id_insegnante, 
    docenti.nome, 
    docenti.cognome
    order by numero_corsi_prof desc limit 1
"""

cursor.execute(query13)
nprof = cursor.fetchall()

for id_insegnante, nome, cognome, numero_corsi_prof in nprof:
    print(nome, cognome, numero_corsi_prof)
```

```python
cursor = conn.cursor(buffered=True)
query14 = """
SELECT 
    studenti.nome, 
    studenti.cognome,
    COUNT(studenti_corsi.id) AS stud
FROM studenti
INNER JOIN studenti_corsi
    ON studenti_corsi.id_studenti = studenti.id_studenti
GROUP BY 
    studenti.id_studenti,
    studenti.nome, 
    studenti.cognome
ORDER BY stud DESC
LIMIT 3
"""
cursor.execute(query14)
top3 = cursor.fetchall()

for nome, cognome, stud in top3:
    print(nome, cognome, stud)
```
