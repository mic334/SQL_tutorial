```python
import mysql.connector

conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password=""
)

cursor = conn.cursor()

with open("../create_table.sql", "r", encoding="utf-8") as file:
    sql_script = file.read()

queries = sql_script.split(";")

for i, query in enumerate(queries, start=1):
    query = query.strip()
    if query:
        print(f"\nQUERY {i}:")
        print(query)
        cursor.execute(query)

conn.commit()

print("SQL eseguito ✅")

conn.close()
```

```python
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="",
    database="esercizi_scuola_sql"
)

cursor = conn.cursor()
cursor.execute("SELECT * FROM studenti;")
studenti=cursor.fetchall()
cursor.execute("SELECT * FROM corsi;")
corsi=cursor.fetchall()
cursor.execute("SELECT * FROM docenti;")
docenti=cursor.fetchall()

# risultati
for riga in studenti:
    print(riga)

for riga in corsi:
    print(riga)

for riga in docenti:
    print(riga)
conn.close()
```

```python
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="",
    database="esercizi_scuola_sql"
)


cursor = conn.cursor()

query = """
UPDATE corsi
SET data_inizio = %s, data_fine = %s, costo = %s
WHERE nome_corso = %s
"""

dati = [
    ("2026-04-01", "2026-06-01", 150, "MATEMATICA"),
    ("2026-05-01", "2026-07-01", 120, "FISICA"),
    ("2026-06-01", "2026-08-01", 100, "CHIMICA"),
    ("2026-06-01", "2026-08-01", 100, "BIOLOGIA"),
    ("2026-06-01", "2026-08-01", 200, "ITALIANO"),
    ("2026-06-01", "2026-08-01", 100, "STORIA")
]

for riga in dati:
    cursor.execute(query, riga)

conn.commit()
```
