```python
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="",
    database="esercizi_scuola_sql"
)

cursor = conn.cursor()


#query 1 select nome cogmoe docenti e print
query1="""SELECT nome, cognome from docenti """
cursor.execute(query1)
docenti = cursor.fetchall()
for riga_docenti in docenti:
    print(riga_docenti)
```

```python
#query 2 email studeni maggiorenni
query2=""" SELECT nome, cognome from studenti
WHERE TIMESTAMPDIFF(YEAR, data_nascita, CURDATE()) >= 22"""
eta = cursor.fetchall()
cursor.execute(query2)

for riga_eta in eta :
    print(riga_eta)
```

```python
query3= """SELECT nome_corso
    FROM corsi
    WHERE data_fine < CURDATE()"""

corsi_scaduti = cursor.fetchall()
cursor.execute(query3)


for riga_fine in corsi_scaduti:
    print(riga_fine)
```

```python
query4="""SELECT nome_corso 
        FROM corsi
        WHERE data_inizio > CURDATE()"""
    
corsi_da_iniziare = cursor.fetchall()
cursor.execute(query4)

for riga_inizio in corsi_da_iniziare:
    print(riga_inizio)
```

```python
query5 = """SELECT nome_corso 
    from corsi 
    WHERE data_inizio > CURDATE() 
    AND
    costo < 150
    """
    
corsi_economico_nuovo=cursor.fetchall()
cursor.execute(query5)

for righe_eco in corsi_economico_nuovo:
    print(righe_eco)
```
