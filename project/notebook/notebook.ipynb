```python
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="",
    database="esercizi_scuola_sql"
)

cursor = conn.cursor()
cursor.execute("SELECT * FROM studenti;")
studenti=cursor.fetchall()
cursor.execute("SELECT * FROM corsi;")
corsi=cursor.fetchall()
cursor.execute("SELECT * FROM docenti;")
docenti=cursor.fetchall()

# risultati
for riga in studenti:
    print(riga)

for riga in corsi:
    print(riga)

for riga in docenti:
    print(riga)
conn.close()
```
